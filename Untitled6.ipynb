```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Путь к файлу
file_path = '/content/photo_2024-12-30_12-49-33.jpg'

# Чтение изображения
image = cv2.imread(file_path)
if image is None:
    print("Ошибка: Не удалось загрузить изображение. Проверьте путь и имя файла.")
else:
    # Уменьшение размера изображения
    scale_percent = 50
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    dim = (width, height)
    image = cv2.resize(image, dim)

    # Вывод исходного изображения
    print("Исходное изображение:")
    cv2_imshow(image)

    # Преобразование в оттенки серого
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    print("Изображение в оттенках серого:")
    cv2_imshow(gray)

    # Применение пороговой обработки (метод Оцу)
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    print("Пороговое изображение:")
    cv2_imshow(thresh)

    # Морфологические операции для удаления шумов
    kernel = np.ones((3, 3), np.uint8)  # Размер ядра
    thresh = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)  # Удаление шумов

    # Нахождение контуров (RETR_LIST)
    contours, _ = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

    # Фильтрация контуров по площади и удаление контуров на границах
    min_area = 1200  # Минимальная площадь контура
    border_margin = 5  # Отступ от границы изображения
    filtered_contours = []
    for contour in contours:
        # Проверка, что контур не касается границ изображения
        x, y, w, h = cv2.boundingRect(contour)
        if (x > border_margin and y > border_margin and
            x + w < width - border_margin and
            y + h < height - border_margin and
            cv2.contourArea(contour) > min_area):
            filtered_contours.append(contour)

    # Проверка, что контуры найдены
    if not filtered_contours:
        print("Контуры не найдены.")
    else:
        # Вывод количества объектов
        num_objects = len(filtered_contours)
        print(f"Количество объектов на изображении: {num_objects}")

        # Создание копии изображения для отрисовки контуров
        image_with_contours = image.copy()

        # Сортировка контуров по площади
        filtered_contours = sorted(filtered_contours, key=cv2.contourArea, reverse=True)

        # Нахождение центров объектов
        centers = []
        for contour in filtered_contours:
            M = cv2.moments(contour)
            if M['m00'] != 0:  # Проверка, чтобы избежать деления на ноль
                cx = int(M['m10'] / M['m00'])
                cy = int(M['m01'] / M['m00'])
                centers.append((cx, cy))

        # Проверка, что центры найдены
        if not centers:
            print("Не удалось вычислить центры объектов.")
        else:
            # Обведение контуров и центров
            for i, contour in enumerate(filtered_contours):
                cv2.drawContours(image_with_contours, [contour], -1, (0, 255, 0), 2)
                if i < len(centers):  # Проверка, чтобы избежать выхода за пределы списка
                    cv2.circle(image_with_contours, centers[i], 5, (0, 0, 255), -1)

            # Вывод информации о центрах
            if len(centers) >= 1:
                cv2.putText(image_with_contours, f'Biggest: {centers[0]}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
            if len(centers) >= 2:
                cv2.putText(image_with_contours, f'Smallest: {centers[-1]}', (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

            # Вывод количества объектов на изображении
            cv2.putText(image_with_contours, f'Objects: {num_objects}', (10, 110), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

            # Вывод изображения с контурами и центрами
            print("Изображение с контурами и центрами объектов:")
            cv2_imshow(image_with_contours)

            # Сохранение результата
            cv2.imwrite('result.jpg', image_with_contours)
```
